```python
import numpy as np
import pandas as pd
import torch

from torch import nn
m = nn.ReLU()

# audio player
import IPython.display as ipd
import scipy
```

```python
genres = ['blues', 'classical', 'country', 'disco', 'hiphop', 'jazz', 'metal', 'pop', 'reggae', 'rock']
```

```python
def gen_random_1d_vector(size):
    return torch.randn(size)

vec2 = gen_random_1d_vector(2048)
vec2.shape
```

```python
song_embeddings = []
song_names = []

for genre in genres:
    for num in range(100):
        num_str = str(num)
        num_str = "000" + num_str
        if len(num_str) == 4:
            num_str = '0' + num_str
        filename = f'./Data/genres_original/{genre}/{genre}.{num_str}.wav'
        # embedding = yaremas_magical_func(filename)
        embedding = gen_random_1d_vector(2048)
        song_embeddings.append(np.array(embedding))
        song_names.append(filename)

song_embeddings = np.array(song_embeddings)
song_embeddings.shape
```

```python
distance_matrix = scipy.spatial.distance.cdist(song_embeddings, song_embeddings, metric='euclidean') # cosine
```

```python
# init the embedding matrix
embedding_matrix = [None for _ in range(len(song_embeddings))]

song_counter = 0
for i in range(len(embedding_matrix)):
    embedding_matrix[i] = [[distance_matrix[i][j], song_names[j], j] for j in range(len(song_embeddings))]
```

```python
embedding_matrix[0][1]
```

```python
def get_recommendations(song_name):
    j = song_names.index(song_name)
    distances = embedding_matrix[j]
    distances.sort(key=lambda x: x[0])
    # getting the top 5 recommendations:
    top5 = distances[1:6]
    return top5

recs = get_recommendations('./Data/genres_original/rock/rock.00010.wav')
recs_names = [song[1] for song in recs]
recs_names
recs
```
